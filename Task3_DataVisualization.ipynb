```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("books_data.csv")
print("Dataset loaded successfully with", len(df), "rows")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=20, kde=True, color='lightgreen')
plt.title("Price Distribution of Books")
plt.xlabel("Price (£)")
plt.ylabel("Number of Books")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='rating', data=df, color='orange')
plt.title("Books by Rating")
plt.xlabel("Rating (Stars)")
plt.ylabel("Number of Books")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='rating', y='price', data=df, palette="Blues")
plt.title("Average Price by Rating")
plt.xlabel("Rating (Stars)")
plt.ylabel("Average Price (£)")
plt.show()
```

```python
top_10 = df.sort_values(by='price', ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x='price', y='title', data=top_10, palette="cool")
plt.title("Top 10 Most Expensive Books")
plt.xlabel("Price (£)")
plt.ylabel("Book Title")
plt.show()
```

```python
rating_counts = df['rating'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Distribution of Book Ratings")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='rating', y='price', data=df, palette="Set2")
plt.title("Price Distribution by Rating")
plt.xlabel("Rating (Stars)")
plt.ylabel("Price (£)")
plt.show()
```
